```python
# Import libararies
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

### If there is no error message, you can start plotting cities with GeoPandas here:

```python
# Answer (plot citis with LAT and LON values):

df = pd.DataFrame(
    {
        "City": ["Chicago", "Houston", "Austin", "New York City", "Cape Canaveral"],
        "Longitude": [-87.6298, -95.3698, -97.7431, -74.0060, -80.6077],
        "Latitude": [41.8781, 29.7604, 30.2672, 40.7128, 28.3922],
    }
)

fig, ax = plt.subplots(1, 1, figsize=(16,8))

gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Longitude, df.Latitude))
gdf.plot(ax=ax, color="red", markersize=100)

states = gpd.read_file("./states/cb_2018_us_state_500k.shp").to_crs("EPSG:4326")
ignore_states = ["Alaska", "Hawaii", "Puerto Rico", "United States Virgin Islands", "Commonwealth of the Northern Mariana Islands", "Guam", "American Samoa"]
states[~states["NAME"].isin(ignore_states)].boundary.plot(ax=ax)
```

### 2a. What is SDI and what are the goals of SDI? (2pt)

```python
'''
Spatial Data Infrastructure (SDI) is a framework of policies, institutional 
arrangements, technologies, data, and people which allows for effective sharing 
and use of geographic data through standardization of formats and protocols.

The goals of SDI are to reduce duplication of efforts, lower costs related to 
geographic information and make data more accessible, increase the benefits of 
using available spatial data, and establish partnerships between states, 
counties, cities, academia, and the private sector.
'''
```

### 2b. Why SDI is important? (3pt)

```python
'''
SDI is important in order to reduce wasted efforts. Currently, many people and 
organizations at the state, country, city, academia, and private sector levels 
are duplicating each other's work. By standardizing data formats and protocols, 
one organization can complete the work, and share the benefits with everyone 
else. Some downsides to this would be agreeing upon the standards. For example, 
when considering the topic of spatial projections, many people have their 
preferences and it will depend on the use case. Not everyone may be able to 
agree to the standards. That is why INSPIRE is also important -- it provides a 
legal framework for developing SDI (focused in the European Union). If the 
government enforces these standards, people will be extremely likely to follow 
them. I hope the United States can create a similar legal framework. Or, maybe
we can follow in the footsteps and standards of corporations such as Google,
who have begun to develop many GIS products.
'''
```

```python
'''
There are many benefits to cloud computing. It is generally cheaper to use for 
users, it is fast, even on computers like Chromebooks, it can scale extremely 
well, increase productivity by allowing users to code instead of deal with 
hardware. It is extremely reliable, with downtimes measured in milliseconds per 
year. It is secure, users do not have to worry about being hacked and losing 
valuable data, since it is the responsbility of the provider.

There are three types of clouds. The most common for small-scale users is the 
public cloud. These are owned and operated by third-party providers, delivering 
the computation results over the Internet. Examples of public cloud include 
Amazon Web Services and Microsoft Azure. A private cloud is used exclusively by 
a single business or organization. These can be physically located in the 
company's datacenter. These are private and communications typically never 
reach the internet. The final type of cloud is hybrid cloud, which combines 
public and private cloud to allow data and applications to be shared between 
them. This is a good choice for a company that is trying to transition from 
on-premises hardware to a 100% cloud solution.
'''
```
